## Preparing Retention dataset for binomial regression

```python
'''
I think a better dataset would have been to have different columns denoting the features of individual students, and 
then a last column denoting the semester you retained to. So classification problem using logistic regression?
Also, should we have excluded years 2018 and 2019 because we don't have their all 3 semesters retention rates?
'''
```

```python
Ret[191:192].to_dict()
```

```python
newDF = pd.DataFrame()
GENDER = []
BIPOCFmt = []
FGFmt = []
PellFmt = []
Group2 = []
EYEAR = []
FallSemester = []
Retained_or_not = []

for i in range(192):
    dictionary = Ret[i:i+1].to_dict()
    for j in ['Ret2nd', 'Ret3rd', 'Ret4th']:
        if dictionary[j][i] == dictionary[j][i]: # Not a nan value
            Number_of_total_individuals = dictionary['N'][i]
            Retention_rate = dictionary[j][i]
            Number_of_retained_individuals = round(Retention_rate*Number_of_total_individuals)
            Number_of_notRetained_individuals = Number_of_total_individuals - Number_of_retained_individuals
            for k in range(Number_of_retained_individuals):
                GENDER.append(dictionary['GENDER'][i])
                BIPOCFmt.append(dictionary['BIPOCFmt'][i])
                FGFmt.append(dictionary['FGFmt'][i])
                PellFmt.append(dictionary['PellFmt'][i])
                Group2.append(dictionary['Group2'][i])
                EYEAR.append(dictionary['EYEAR'][i])
                FallSemester.append(j)
                Retained_or_not.append(1)
            for k in range(Number_of_notRetained_individuals):
                GENDER.append(dictionary['GENDER'][i])
                BIPOCFmt.append(dictionary['BIPOCFmt'][i])
                FGFmt.append(dictionary['FGFmt'][i])
                PellFmt.append(dictionary['PellFmt'][i])
                Group2.append(dictionary['Group2'][i])
                EYEAR.append(dictionary['EYEAR'][i])
                FallSemester.append(j)
                Retained_or_not.append(0)
            
print(len(GENDER), len(BIPOCFmt), len(FGFmt), len(PellFmt), len(Group2), len(EYEAR), len(FallSemester), len(Retained_or_not
           
newDF['Gender'] = GENDER
newDF['BIPOC_or_not'] = BIPOCFmt
newDF['FirstGen_or_not'] = FGFmt
newDF['PellGrant'] = PellFmt
newDF['Rural_or_not'] = Group2
newDF['EntryYear'] = EYEAR
newDF['Fall_Sem'] = FallSemester
newDF['Retained_or_not'] = RetentionPercent
```

```python
newDF
```

```python
# I think we should stick to 2014 to 2017 only, because for 2018 and 2019 we do not have data for 3rd and 4th Fall.
# So it won't be nonsensical to interpret the odds of retention of 2019 students to their 4th Fall.
Data = Data[Data.EYEAR <= 2017]
Data.reset_index(drop = False, inplace = True)
```
