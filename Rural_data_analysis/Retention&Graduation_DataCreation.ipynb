```python
def test_of_proportions(p1_hat, N, p2_hat, M):
    print("Rural percentage =", str(round(p1_hat*100, 1))+str("%"), ", Non-rural percentage = ", str(round(p2_hat*100,1))+str("%"))
    n = int(p1_hat*N)
    m = int(p2_hat*M)
    p_hat = (n+m)/(N+M)
    q1_hat = 1 - p1_hat
    q2_hat = 1 - p2_hat
    q_hat = 1 - p_hat
    print(p1_hat*m>=10, q1_hat*m>=10, p2_hat*n>=10, q2_hat*n>=10)
    z = (p1_hat - p2_hat)/math.sqrt(p_hat*q_hat*(1/N + 1/M))
    print("test statistic = ", z)
    print(stats.norm.cdf(z))
    print(stats.norm.ppf(0.05))
    print(stats.norm.ppf(1-0.05))
    if z < stats.norm.ppf(0.05): #or z > stats.norm.ppf(1-0.05/2):
        print("Statistically significant.\n")
    else:
        print("Not statistically significant.\n")
```
