```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
RententionData = pd.read_csv('../RuralData/RetentionData_Final.csv')
RententionData
```

```python
# Plot 1: Gender VS RetentionRate, Legend = Rural_or_not
DF1 = RententionData[(RententionData.Gender == "F") & (RententionData.Rural_or_not == 1)]
Ret_RuralFemale = sum(DF1['Retained_or_not'])/len(DF1['Retained_or_not'])*100
print(Ret_RuralFemale)

DF2 = RententionData[(RententionData.Gender == "M") & (RententionData.Rural_or_not == 1)]
Ret_RuralMale = sum(DF2['Retained_or_not'])/len(DF2['Retained_or_not'])*100
print(Ret_RuralMale)

DF3 = RententionData[(RententionData.Gender == "F") & (RententionData.Rural_or_not == 0)]
Ret_NonRuralFemale = sum(DF3['Retained_or_not'])/len(DF3['Retained_or_not'])*100
print(Ret_NonRuralFemale)

DF4 = RententionData[(RententionData.Gender == "M") & (RententionData.Rural_or_not == 0)]
Ret_NonRuralMale = sum(DF4['Retained_or_not'])/len(DF4['Retained_or_not'])*100
print(Ret_NonRuralMale)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
Genders = ["Male", "Female"]
plt.plot(Genders, [Ret_RuralMale, Ret_RuralFemale], "go-", label = "Rural")
plt.plot(Genders, [Ret_NonRuralMale, Ret_NonRuralFemale], "ro-", label = "Non-Rural")
plt.xlabel("Gender", fontsize = 15, color = "black")
plt.ylabel("Retention rate", fontsize = 15, color = "black")
fmt = '%2.0f%%' # Format you want the ticks, e.g. '40%'
yticks = mtick.FormatStrFormatter(fmt)
ax.yaxis.set_major_formatter(yticks)

plt.xticks(fontsize = 15, color = "black")
plt.yticks(fontsize = 15, color = "black")
plt.legend(fontsize = 15)
plt.show()
```

```python
# Plot 2: BIPOC VS RetentionRate, Legend = Rural_or_not
DF1 = RententionData[(RententionData.BIPOC_or_not == 1) & (RententionData.Rural_or_not == 1)]
Ret_BIPOCRural = sum(DF1['Retained_or_not'])/len(DF1['Retained_or_not'])*100
print(Ret_BIPOCRural)

DF2 = RententionData[(RententionData.BIPOC_or_not == 0) & (RententionData.Rural_or_not == 1)]
Ret_NonBIPOCRural = sum(DF2['Retained_or_not'])/len(DF2['Retained_or_not'])*100
print(Ret_NonBIPOCRural)

DF3 = RententionData[(RententionData.BIPOC_or_not == 1) & (RententionData.Rural_or_not == 0)]
Ret_BIPOCNonRural = sum(DF3['Retained_or_not'])/len(DF3['Retained_or_not'])*100
print(Ret_BIPOCNonRural)

DF4 = RententionData[(RententionData.BIPOC_or_not == 0) & (RententionData.Rural_or_not == 0)]
Ret_NonBIPOCNonRural = sum(DF4['Retained_or_not'])/len(DF4['Retained_or_not'])*100
print(Ret_NonBIPOCNonRural)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
BIPOCStatus = ["BIPOC", "Non-BIPOC"]
plt.plot(BIPOCStatus, [Ret_BIPOCRural, Ret_NonBIPOCRural], "go-", label = "Rural")
plt.plot(BIPOCStatus, [Ret_BIPOCNonRural, Ret_NonBIPOCNonRural], "ro-", label = "Non-Rural")
plt.xlabel("BIPOC Status", fontsize = 15, color = "black")
plt.ylabel("Retention rate", fontsize = 15, color = "black")
fmt = '%2.0f%%' # Format you want the ticks, e.g. '40%'
yticks = mtick.FormatStrFormatter(fmt)
ax.yaxis.set_major_formatter(yticks)

plt.xticks(fontsize = 15, color = "black")
plt.yticks(fontsize = 15, color = "black")
plt.legend(fontsize = 15)
plt.show()

# This might be a reason why BIPOC term in the model is non-significant, because if you fix rural, there is decrease,
# but if you fix non-rural, then increase, so on average, there is no change.
```

```python
# Plot 3: PellGrant VS RetentionRate, Legend = Rural_or_not
DF1 = RententionData[(RententionData.PellGrant == 1) & (RententionData.Rural_or_not == 1)]
Ret_PellRural = sum(DF1['Retained_or_not'])/len(DF1['Retained_or_not'])*100
print(Ret_PellRural)

DF2 = RententionData[(RententionData.PellGrant == 0) & (RententionData.Rural_or_not == 1)]
Ret_NonPellRural = sum(DF2['Retained_or_not'])/len(DF2['Retained_or_not'])*100
print(Ret_NonPellRural)

DF3 = RententionData[(RententionData.PellGrant == 1) & (RententionData.Rural_or_not == 0)]
Ret_PellNonRural = sum(DF3['Retained_or_not'])/len(DF3['Retained_or_not'])*100
print(Ret_PellNonRural)

DF4 = RententionData[(RententionData.PellGrant == 0) & (RententionData.Rural_or_not == 0)]
Ret_NonPellNonRural = sum(DF4['Retained_or_not'])/len(DF4['Retained_or_not'])*100
print(Ret_NonPellNonRural)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
Genders = ["Pell Grant Recepient", "Non Pell Grant Recepient"]
plt.plot(Genders, [Ret_PellRural, Ret_NonPellRural], "go-", label = "Rural")
plt.plot(Genders, [Ret_PellNonRural, Ret_NonPellNonRural], "ro-", label = "Non-Rural")
plt.xlabel("Pell Grant Status", fontsize = 15, color = "black")
plt.ylabel("Retention rate", fontsize = 15, color = "black")
fmt = '%2.0f%%' # Format you want the ticks, e.g. '40%'
yticks = mtick.FormatStrFormatter(fmt)
ax.yaxis.set_major_formatter(yticks)

plt.xticks(fontsize = 15, color = "black")
plt.yticks(fontsize = 15, color = "black")
plt.legend(fontsize = 15)
plt.show()

#Email Karen ans ask explanation
```

There is no crossing of lines here, but the interaction term (one of the 4 int. terms) in the model is significant. It is not surprising because this is an overall plot, in the model we are controlling for many other predictors. If we take only rural and Pell in a model, we see the int. term is not significant. So that matches with this plot.

```python
# Plot 3: PellGrant VS RetentionRate, Legend = Rural_or_not
DF1 = RententionData[(RententionData.PellGrant == 1) & (RententionData.Rural_or_not == 1) & (RententionData.Gender == "M") ]
Ret_PellRuralMale = sum(DF1['Retained_or_not'])/len(DF1['Retained_or_not'])*100
print(Ret_PellRuralMale)

DF11 = RententionData[(RententionData.PellGrant == 1) & (RententionData.Rural_or_not == 1) & (RententionData.Gender == "F") ]
Ret_PellRuralFemale = sum(DF11['Retained_or_not'])/len(DF11['Retained_or_not'])*100
print(Ret_PellRuralFemale)

DF2 = RententionData[(RententionData.PellGrant == 1) & (RententionData.Rural_or_not == 0) & (RententionData.Gender == "M") ]
Ret_PellNonRuralMale = sum(DF2['Retained_or_not'])/len(DF2['Retained_or_not'])*100
print(Ret_PellNonRuralMale)

DF22 = RententionData[(RententionData.PellGrant == 1) & (RententionData.Rural_or_not == 0) & (RententionData.Gender == "F") ]
Ret_PellNonRuralFemale = sum(DF22['Retained_or_not'])/len(DF22['Retained_or_not'])*100
print(Ret_PellNonRuralFemale)


DF1 = RententionData[(RententionData.PellGrant == 0) & (RententionData.Rural_or_not == 1) & (RententionData.Gender == "M") ]
Ret_NonPellRuralMale = sum(DF1['Retained_or_not'])/len(DF1['Retained_or_not'])*100
print(Ret_NonPellRuralMale)

DF11 = RententionData[(RententionData.PellGrant == 0) & (RententionData.Rural_or_not == 1) & (RententionData.Gender == "F") ]
Ret_NonPellRuralFemale = sum(DF11['Retained_or_not'])/len(DF11['Retained_or_not'])*100
print(Ret_NonPellRuralFemale)

DF2 = RententionData[(RententionData.PellGrant == 0) & (RententionData.Rural_or_not == 0) & (RententionData.Gender == "M") ]
Ret_NonPellNonRuralMale = sum(DF2['Retained_or_not'])/len(DF2['Retained_or_not'])*100
print(Ret_NonPellNonRuralMale)

DF22 = RententionData[(RententionData.PellGrant == 0) & (RententionData.Rural_or_not == 0) & (RententionData.Gender == "F") ]
Ret_NonPellNonRuralFemale = sum(DF22['Retained_or_not'])/len(DF22['Retained_or_not'])*100
print(Ret_NonPellNonRuralFemale)


fig = plt.figure()
ax = fig.add_subplot(1,1,1)
Pell = ["Pell Grant Recepient", "Non Pell Grant Recepient"]
plt.plot(Pell, [Ret_PellNonRuralFemale, Ret_NonPellNonRuralFemale], "ro-" , linestyle = ":", lw = 3, label = "Non-Rural Female")
plt.plot(Pell, [Ret_PellRuralFemale, Ret_NonPellRuralFemale], "go-", linestyle = ":", lw = 3, label = "Rural Female")
plt.plot(Pell, [Ret_PellNonRuralMale, Ret_NonPellNonRuralMale], "ro-", label = "Non-Rural Male")
plt.plot(Pell, [Ret_PellRuralMale, Ret_NonPellRuralMale], "go-", label = "Rural Male")

plt.xlabel("Pell Grant Status", fontsize = 15, color = "black")
plt.ylabel("Retention rate", fontsize = 15, color = "black")
fmt = '%2.0f%%' # Format you want the ticks, e.g. '40%'
yticks = mtick.FormatStrFormatter(fmt)
ax.yaxis.set_major_formatter(yticks)

#plt.ylim(75, 100)
#plt.plot(["Pell Grant Recepient"],80, markersize = 5, linestyle = ":", lw = 3, "")
plt.xticks(fontsize = 15, color = "black")
plt.yticks(fontsize = 15, color = "black")
plt.legend(fontsize = 15)
plt.show()

# Its not that among Females the lines cross and among males it doesn't and so gender explains. There must be 
# something else due to which the int term between Pell and rural came out to be sig, but not with gender!!
```
