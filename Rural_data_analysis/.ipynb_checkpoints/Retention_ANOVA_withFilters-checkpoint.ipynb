```r
RententionData = read.csv(file = '../RuralData/Retention_wFilters.csv')
head(RententionData)
RententionData
```

```r
RententionData$GENDER = as.factor(RententionData$GENDER)
RententionData$BIPOCFmt = as.factor(RententionData$BIPOCFmt)
RententionData$FGFmt = as.factor(RententionData$FGFmt)
RententionData$PellFmt = as.factor(RententionData$PellFmt)
RententionData$Group2 = as.factor(RententionData$Group2)
RententionData$EYEAR = as.factor(RententionData$EYEAR)
RententionData$FallSemester = as.factor(RententionData$FallSemester)

head(RententionData)
```

```r
lmod = lm(RetentionPercent ~ GENDER + BIPOCFmt + FGFmt + PellFmt + Group2 + FallSemester, data = RententionData)
summary(lmod)
```

```r
lmod = lm(RetentionPercent ~ Group2 + FallSemester, data = RententionData)
summary(lmod)
```

We see that the rural and non-rural do not have a statistically significant difference in Retention rate using this 
data, although using the overall .csv file we HAD found a statistically significance difference (difference across
semesters was significant there too, like we have here). This difference is most probably because of this: Note that
the following output is saying that on average, the difference between retention rate of rural and non-rural
students in the 2nd Fall semester is 0.6%, but our summary using the overall .csv file showed that this difference
was 2.2% for 2nd Fall and was statistically significant. The reason these two percentages do not match is that 
in this csv, the percentages are grouped, and thats a problem. Suppose you have 110 rural students in total. 10 of 
them are female rural students and 100 of them are male rural students. Lets also suppose that the retention rate of
the female group is 90%, meaning 9 of 10 females retained, and that of the male is 100%, meaning all males retained.
Then as per the above grouped csv file format, the average rentention rate of the rural students comes out to be 95%
because there is one row saying 90% and another saying 100%, so average is 95%. But, if this same data is in the 
format of the overall .csv file, then the average retention rate of all the rural students would be printed as 99.1%.
Notice the difference? This is why numbers in these two models for the same baseline group is also not matching in 
the first place, let alone the difference between rural and non-rural groups. I dont think there is a straight
solution solution to this issue (but glad that I identified it) except for changing this dataset such that we have a 
column of 0/1 indicating whether the student retained or not, and of course corresponding columns denoting the
features of the student. And then we can apply binomial(or bornoulli) regression and look at the odds of success, 
where success means retention.

```r
lmod = lm(RetentionPercent ~ Group2, data = RententionData)
summary(lmod)
```
